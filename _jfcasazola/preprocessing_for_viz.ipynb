Al ver que la columna "date" no tenía el dtype que debería (datetime), por lo que lo cambiamos.

Y cuando tengamos la columna "date" con el dtype correcto tendremos que sacar su propio "yearweek", los nulos de la columna "event" serán tratados más adelante

```python
df_dcwe["date"] = pd.to_datetime(df_dcwe["date"], format="%Y-%m-%d")
df_dcwe["yearweek"] = df_dcwe["date"].dt.strftime("%Y%W").astype(float)

df_dcwe.info()
```

```python
list_days = [x for x in df_itsa.columns.to_list() if x.startswith("d_")]
list_columns = [x for x in df_itsa.columns.to_list() if not x.startswith("d_")]

df_sales = pd.melt(df_itsa, id_vars=list_columns, value_vars=list_days, 
                   var_name="d", value_name="n_sales")
```

## Merging DFs

### Merge de Item Prices con Daily Calendar with Events

```python
df_merged = pd.merge(df_itpr, 
                     df_dcwe[["date","d","yearweek", "event"]], 
                     how="left",
                     on="yearweek")
```

### Merge de df_merged con df_sales

```python
df_completo = pd.merge(df_sales[["id","item","category","department","store_code",
                                "region","d","n_sales"]],
                    df_merged[["item","category","store_code","yearweek","date",
                                "d","sell_price", "event"]], 
                    how="left",
                    on=["item","category","store_code","d"])
```

```python
moreInfo(df_completo)
```

```python
df_completo
```

## Últimas transformaciones

Al ver que nuestro dataframe tenía date con nulos (cuando deberían tener un valor relacionado con la columna "d") pues decidimos crear un dataframe con el que podríamos conseguir las fechas faltantes y sustituir la columna "date" original por una que no tuviera nulos y se basara en la columna "d".

```python
list_un_d = df_completo["d"].unique()

dates = pd.date_range(start=df_completo["date"].min(), 
                      periods=len(list_un_d), 
                      freq="D")

date_mapping = pd.DataFrame({"d": list_un_d,"date": dates})

df = pd.merge(df_completo[[
    "id",
    "item",
    "category",
    "department",
    "store_code",
    "region",
    "event",
    "d",
    "n_sales",
    "sell_price"
 ]], 
 date_mapping, on="d", how="left")

df.drop("d", axis=1, inplace=True)
```

```python
df
```

Aun nos quedaban unos problemas pendientes y eran rellenar los nulos en la columna "event" y crear una columna para poder graficar las ganancias en una visualización, la cual cálculamos multiplicando el precio del producto por el número de ventas.

```python
df["event"].fillna("none", inplace=True)
df["sell_price"] = df.groupby("id")["sell_price"].bfill()
df["sell_price"] = round(df["sell_price"], 2)
df["revenues"] = round(df["sell_price"] * df["n_sales"], 2)
```
